```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Step1: Load Data 

```python
data = pd.read_csv("../Excel/diabetes.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
class_0 = data[data.Outcome==0].sample(268)
class_1 = data[data.Outcome==1]
data = pd.concat((class_0, class_1)).sample(536)
```

# Step2: Separate Input and out features

```python
X = data.drop('Outcome',axis=1)
y = data['Outcome']
```

# Step3: Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0, test_size=.20)
```

# Common Classification Algorithms:
# Logistic Regression:
                A statistical method for predicting the probability of an event occurring. 
# Support Vector Machines (SVM):
                A powerful algorithm that finds the optimal hyperplane to separate data points into different classes. 
# Decision Trees:
                Algorithms that use a tree-like structure to make predictions based on a series of decisions. 
# Random Forest:
                An ensemble method that combines multiple decision trees to improve accuracy. 
# K-Nearest Neighbors (KNN):
                An algorithm that classifies data points based on the majority class of their nearest neighbors. 
# Neural Networks:
                Complex models inspired by the structure of the human brain, capable of learning complex patterns. 

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
```

# Step5: Model Fit

```python
logr.fit(X_train,y_train)
```

```python
y_pred_train = logr.predict(X_train)
y_train
df = pd.DataFrame(data = [y_train.values, y_pred_train], index = ["Train Data", "Predicted"])
print("Real Data", y_train.values,"\n \n \n \n", "Predicted", y_pred_train)
df.T.head(20)
```

```python
count = 0
for t,p in zip(y_train,y_pred_train):
    # print(t, p)
   if t==p:
       count+=1
count
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_train, y_pred_train))
```

```python
confusion_matrix(y_train, y_pred_train)
```

```python
y_pred_test = logr.predict(X_test)
confusion_matrix(y_test, y_pred_test)
```

```python
print(classification_report(y_test, y_pred_test))
```
