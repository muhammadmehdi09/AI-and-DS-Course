# Step1: Load Data

```python
import pandas as pd
data = pd.read_csv('../Excel/real_estate_data.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data = data.drop(columns=["No"], axis=1)
data["X1 transaction date"] = data["X1 transaction date"].apply(lambda x: int((str(x).split(".")[0])))
data.head()
```

```python
data.isna().sum()
```

# Step2: Separate input data and output data

```python
X = data.drop('Y house price of unit area',axis=1)
y = data['Y house price of unit area']
```

# Step:3 Splitting Data and Training Model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0, test_size=.30)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
y_pred_train = model.predict(X_train)
# pd.DataFrame({"Actual":y_train.values.reshape(7000,), "Prediction":y_pred_train })
r2_score(y_train, y_pred_train) * 100
```

```python
y_pred_test = model.predict(X_test)
# pd.DataFrame({"Actual":y_test.values.reshape(3000,), "Prediction":y_pred_test })
```

```python
# Testing Data
r2_score(y_test, y_pred_test) * 100
```
