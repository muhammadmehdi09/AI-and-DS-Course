# Step1: Load Data

```python
import pandas as pd
import warnings
data = pd.read_csv('../../Excel/customer_churn_data.csv')
warnings.filterwarnings("ignore")
```

```python
data.shape
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data["gender"] = data["gender"].apply(lambda x: 1 if x == "Male" else 0)
data["Partner"] = data["Partner"].apply(lambda x: 1 if x == "Yes" else 0)
data["Dependents"] = data["Dependents"].apply(lambda x: 1 if x == "Yes" else 0)
data["PhoneService"] = data["PhoneService"].apply(lambda x: 1 if x == "Yes" else 0)
data["MultipleLines"] = data["MultipleLines"].apply(lambda x: 1 if x == "Yes" else 0)
data["OnlineSecurity"] = data["OnlineSecurity"].apply(lambda x: 1 if x == "Yes" else 0)
data["OnlineBackup"] = data["OnlineBackup"].apply(lambda x: 1 if x == "Yes" else 0)
data["InternetService"] = data["InternetService"].apply(lambda x: 1 if x == "DSL" else 0)
data["DeviceProtection"] = data["DeviceProtection"].apply(lambda x: 1 if x == "Yes" else 0)
data["TechSupport"] = data["TechSupport"].apply(lambda x: 1 if x == "Yes" else 0)
data["StreamingTV"] = data["StreamingTV"].apply(lambda x: 1 if x == "Yes" else 0)
data["StreamingMovies"] = data["StreamingMovies"].apply(lambda x: 1 if x == "Yes" else 0)
data["PaperlessBilling"] = data["PaperlessBilling"].apply(lambda x: 1 if x == "Yes" else 0)
data["Churn"] = data["Churn"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
encoded = pd.get_dummies(data[['Contract', 'PaymentMethod']], drop_first=True, dtype=int)
data = pd.concat([data, encoded], axis=1)
data.drop(columns=["Contract", "PaymentMethod", "customerID"], inplace=True, axis=1)
data.head()

# encoded = pd.get_dummies(data[['Contract', 'PaymentMethod', 'InternetService']], drop_first=True, dtype=int)
# data = pd.concat([data, encoded], axis=1)
# data.drop(columns=["Contract", "PaymentMethod", "customerID", "InternetService"], inplace=True, axis=1)
# data.head()
```

# Step2: Separate input data and output data

```python
X = data.drop('Churn',axis=1)
y = data['Churn']
```

# Step:3 Splitting Data and Training Model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0, test_size=.30)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
```
