# Step2: Separate Input and out features

```python
X = data.drop('target',axis=1)
y = data['target']
```

# Step3: Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0, test_size=.30)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
```

# Step5: Model Fit

```python
logr.fit(X_train,y_train)
```

```python
y_pred_train = logr.predict(X_train)
# y_train
# df = pd.DataFrame(data = [y_train.values, y_pred_train], index = ["Train Data", "Predicted"])
# print("Real Data", y_train.values,"\n \n \n \n", "Predicted", y_pred_train)
# df.T.head(50)
```

```python
count = 0
for t,p in zip(y_train,y_pred_train):
   if t==p:
       count+=1

count/len(y_train) * 100
```

```python
y_pred_test = logr.predict(X_test)
```

```python
count = 0
for t,p in zip(y_test,y_pred_test):
   if t==p:
       count+=1
       
count/len(y_test) * 100
```
