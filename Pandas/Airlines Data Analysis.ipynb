```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data: categories and counts
data1 = {
    'Issue': data[data["Inflight entertainment"] < 4]["Age Band"].value_counts().index,
    'Total': data[data["Inflight entertainment"] < 4]["Age Band"].value_counts().values
}

df = pd.DataFrame(data1)
df = df.sort_values(by='Total', ascending=False)  # Sort descending

# Calculate cumulative percentage
df['Cumulative %'] = df['Total'].cumsum() / df['Total'].sum() * 100

# Create bar chart for count
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(df['Issue'], df['Total'], color='skyblue')

# Create line chart for cumulative %
ax2 = ax1.twinx()
ax2.plot(df['Issue'], df['Cumulative %'], color='red', marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel('Cumulative Percentage')

# Plot settings
plt.title('Pareto Chart of Customer Issues')
plt.grid(axis='y', alpha=0.7)
plt.tight_layout()
plt.show()
```
