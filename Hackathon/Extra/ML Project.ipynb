# Step1: Load Data

```python
import pandas as pd
data = pd.read_csv('advertising.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.describe()
```

# Step2: Separate input data and output data

```python
X = data[['TV', 'Radio', 'Newspaper']]
y = data[['Sales']]
```

# Step3: Split Data into Training data and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test   = train_test_split(X,y,random_state=0, test_size=.20)
```

# Step:4 Model Selection

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

# Step:5 Model Training

```python
model.fit(X_train,y_train)
```

# Step6: Prediction on Training Data

```python
y_pred_train = model.predict(X_train)
```

```python
y_train.values.reshape(160,)
```

```python
y_pred_train.reshape(160,)
```

```python
pd.DataFrame({"y_train":y_train.values.reshape(160,), "y_pred":y_pred_train.reshape(160,) })
```

# Step7: Prediction on Test  Data 

```python
y_pred_test = model.predict(X_test)
```

```python
pd.DataFrame({"y_test":y_test.values.reshape(40,), "y_pred":y_pred_test.reshape(40,) })
```

# Step8: Calculating Accuracy

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# Training Data
r2_score(y_train, y_pred_train)
```

```python
# Testing Data
r2_score(y_test, y_pred_test)
```
